```python
def personne():
  personnes = [] 
  while True:
    
      nom = input("Quel est votre nom ? ")
      prenom = input("Quel est votre prénom ? ")
      genre = input("Quel est votre genre ? ")
      age = int(input("Quel est votre âge ? "))
      personnes.append((prenom, nom, genre, age))
      continuer = input("Voulez-vous ajouter une autre personne ? (oui/non) : ")
      if continuer.lower() != 'oui':
         break  

  return personnes

def lister_personnes(liste_personnes):
  print("\nListe des personnes :")
  for i, (prenom, nom, genre, age) in enumerate(liste_personnes, start=1):
      print(f"{i}. {prenom} {nom}, {genre}, {age} ans")

def chercher_personne(liste_personnes, nom):
  for prenom, nom_personne, genre, age in liste_personnes:
      if nom_personne.lower() == nom.lower():
          return (prenom, nom_personne, genre, age)  
  return None  


liste_de_personnes = personne()  

lister_personnes(liste_de_personnes)

nom_a_chercher = input("\nEntrez le nom de la personne à chercher : ")
resultat = chercher_personne(liste_de_personnes, nom_a_chercher)

if resultat:
  print(f"Personne trouvée : {resultat[0]} {resultat[1]}, {resultat[2]}, {resultat[3]} ans")
else:
  print("Personne non trouvée.")
```

from math import sqrt

```python
from math import sqrt

def saisir_nombres():
    nombres = input("Entrez des nombres réels : ")
    return list(map(float, nombres.split()))

def sum(liste):
    total = 0
    for nombre in liste:
        total += nombre
    return total

def moyenne(liste):
    return sum(liste) / len(liste)

def produit(liste):
    prod = 1
    for nombre in liste:
        prod *= nombre
    return prod

def max(liste):
    max_value = liste[0]
    for nombre in liste:
        if nombre > max_value:
            max_value = nombre
    return max_value

def min(liste):
    min_value = liste[0]
    for nombre in liste:
        if nombre < min_value:
            min_value = nombre
    return min_value

def ecart_type(liste):
    moy = moyenne(liste)
    var = variance(liste, moy)
    return sqrt(var)

def variance(liste, moy=None):
    if moy is None:
        moy = moyenne(liste)
    var = 0
    for nombre in liste:
        var += (nombre - moy) ** 2
    return var / len(liste)

def covariance(liste_x, liste_y):
    moy_x = moyenne(liste_x)
    moy_y = moyenne(liste_y)
    cov = 0
    for x, y in zip(liste_x, liste_y):
        cov += (x - moy_x) * (y - moy_y)
    return cov / len(liste_x)

nombres = saisir_nombres()
print("Somme :", sum(nombres))
print("Moyenne :", moyenne(nombres))
print("Produit :", produit(nombres))
print("Max :", max(nombres))
print("Min :", min(nombres))
print("Écart-type :", ecart_type(nombres))
print("Variance :", variance(nombres))

print("Pour la covariance, entrez deux listes de même longueur.")
nombres_x = saisir_nombres()
nombres_y = saisir_nombres()
print("Covariance :", covariance(nombres_x, nombres_y))
```

```python
def ajouter_element(ensemble,element):
    if element not in ensemble:
        ensemble.add(element)
    return ensemble

mon_ensemble={4,5,6,8}
print(ajouter_element(mon_ensemble,3))

def transformer_liste(liste):
    mon_ensemble=set(liste)
    return mon_ensemble
ma_liste=[3,5,6,7]
print(transformer_liste(ma_liste))

def ensembles_egaux(ensemble1, ensemble2):
    return ensemble1 == ensemble2
ensemble_a={1,3,4,5}
ensemble_b={1,3,4,5}
print("les ensembles a et b sont egaux:",ensembles_egaux(ensemble_a,ensemble_b))

def est_inclus(ensemble1, ensemble2):
    for element in ensemble1:
        if element not in ensemble2:
            return False 
    return True 
ensemble_a={2,3,4,5}
ensemble_b={2,3,4,9,5,8}
print("est ce que l'ensemble a est inclus dans l'ensemble b :",est_inclus(ensemble_a,ensemble_b))

def intersection(ensemble1,ensemble2):
    return ensemble1 & ensemble2
ensemble_a={2,3,4,5}
ensemble_b={3,4,8,4}
print("l'intersection de a et b est:",intersection(ensemble_a,ensemble_b))

def union(ensemble1,ensemble2):
    return ensemble1 | ensemble2
ensemble_a={2,4,5,6}
ensemble_b={4,5,2}
print("l'union de a et b est:",union(ensemble_a,ensemble_b))

def sont_complementaires(ensemble1, ensemble2, ensemble3):
    for element in ensemble3:
        if element not in ensemble1 and element not in ensemble2:
            return False  
    for element in ensemble1:
        if element in ensemble2:
            return False  

    return True 
ensemble_1={1,0,3}
ensemble_2={6,4,5}
ensemble_3={1,2,3,4,5,6}
print("les trois ensembles sont complementaires:",sont_complementaires(ensemble_1,ensemble_2,ensemble_3))
```
